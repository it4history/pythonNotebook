```python
with open('base.csv', 'r') as csvfile:
    csv_base = csv.reader(csvfile)
    
    nord_dict = {}
    years_goriz_list = ['#',]
    for row in csv_base:
        key = row[0].zfill(4)
        if key not in nord_dict:
            nord_dict[key] = []
        if row[1] not in nord_dict[key]:
            nord_dict[key].append(row[1])
        if row[0] not in years_goriz_list:
            years_goriz_list.append(row[0])

ord_dict = collections.OrderedDict(sorted(nord_dict.items()))
years_vert_list = map(list, zip(*[years_goriz_list]))            
            
with open("matrix.csv", "wb") as csv_out:
    writer = csv.writer(csv_out)
    writer.writerows([years_goriz_list])
    for i in years_vert_list[1:]:
        row = []
        row.append(i[0])
        for year in ord_dict:
            row.append(str(len(set(ord_dict[year]) & set(ord_dict[i[0].zfill(4)]))))
        writer.writerows([
                row,
        ])
```
